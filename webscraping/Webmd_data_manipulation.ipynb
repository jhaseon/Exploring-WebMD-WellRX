```python
# Import necessary files 
import pandas as pd
```

```python
# Read Webmd raw csv 
df = pd.read_csv('Webmd_Selenium.csv')
```

```python
# Preview csv file
df
```

```python
# Get all the unique drugs 
unique_names = df['drug'].unique()
print(sorted(unique_names))
```

```python
# Need to get rid of extra Wellbutrin XL oral extractions
# Mistake in scraping
# Original mistake fixed so this won't be needed 
df1 = df.drop_duplicates()
```

```python
# Remove comment columns
# Comments did were not extracted from website
df2 = df1[['drug','effectiveness', 'ease', 'satifaction', 'condition', 'reviewer','date']]
```

```python
# Clean up reviewer and condition values 

def remove_cruft(s):
    return s[10:]
Reviewer = [remove_cruft(s) for s in df1['reviewer']]
Condition = [remove_cruft(s) for s in df1['condition']]
```

```python
# Clean up effectiveness, ease, and satifaction columns 

Effectiveness = df2.effectiveness.str.extract('(\d+)')
Ease = df2.ease.str.extract('(\d+)')
Satisfaction = df2.satifaction.str.extract('(\d+)')
```

```python
# Adding new cleaned columns 

df2['Reviewer'] = Reviewer
df2['Condition'] = Condition
df2['Effectiveness'] = Effectiveness
df2['Ease'] = Ease
df2['Satisfaction'] = Satisfaction 
```

```python
# Drop unedited columns 

df3 = df2.drop(columns = ['effectiveness', 'ease','satifaction', 'reviewer', 'condition'])
```

```python
# Small quality check 
a = df3.groupby(['drug'])
len(a)
```

```python
# Export final Webmd data 
df3.to_csv('Webmd_final.csv', index=False)
```
