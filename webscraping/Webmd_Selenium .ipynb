```python
# Import packages
from selenium import webdriver
import time
import re
import time 
import csv 

# Selenium
# Change your file path to where your chromedriver.exe is  
driver = webdriver.Chrome(r'...\chromedriver_win32\chromedriver.exe') #filepath for chromedriver

# Create an csv file with column names
# Replace 'file name' with your own
with open('Webmd_Selenium' , 'w', encoding = 'utf-8', newline = '') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['drug', 'condition', 'reviewer', 'date', 'effectiveness', 'ease', 'satifaction', 'comment'])
    
    # Will Run Selenium for each URL page 
    for URL in URL_list:
        driver.get(URL)

        # Change scroll down option to 'All Conditions'
        select_elem = driver.find_element_by_id('conditionFilter')
        select_elem.click()
        options = select_elem.find_elements_by_tag_name('option')
        options[len(options)-1].click()

        # Find the total amount of pages per drug 
        reviews = driver.find_elements_by_xpath('//div[@class="userPost"]')
        total_reviews = driver.find_element_by_xpath('//*[@class="totalreviews"]').text
        total_reviews1 = int(re.findall(r'\d+', total_reviews)[0])
        max_pages = total_reviews1 // 5 + 1 

        # Limit the number of times 'Next' button is clicked
        # Selenium will go to next URL when limit is reached
        index = 1 
        while index <= max_pages:
            print(('Scraping page {}'.format(index)) + ' of ' + str(max_pages))
            index += 1 
            
            # Create xpath variable for individual review
            reviews = driver.find_elements_by_xpath('//div[@class="userPost"]')
            for review in reviews:
                
                # Each review will its own dictionary
                # That dictionary will be saved to csv file
                review_dict = {}
                
                # List of elements that we are trying to scrape from WebMD
                try:
                    drug = driver.find_element_by_xpath(".//*[@id='header']").text
                    before_keyword, keyword, after_keyword = drug.partition('- ') # needed to get just the drug name from header
                    condition = review.find_element_by_xpath('.//div[@class="conditionInfo"]').text
                    reviewer = review.find_element_by_xpath('.//p[@class = "reviewerInfo"]').text
                    date = review.find_element_by_xpath('.//div[@class = "date"]').text
                    effectiveness = review.find_element_by_xpath('.//div[@class = "catRatings firstEl clearfix"]').text
                    ease = review.find_element_by_xpath('.//div[@class = "catRatings clearfix"]').text
                    satisfaction = review.find_element_by_xpath('.//div[@class = "catRatings lastEl clearfix"]').text
                    # comments were not extracted because too difficult to extract from xpath
                    
                except:
                    continue
                    
                # Add scraped elements into empty review dictionary 
                review_dict['drug'] = after_keyword
                review_dict['condition'] = condition
                review_dict['reviewer'] = reviewer
                review_dict['date'] = date
                review_dict['effectiveness'] = effectiveness
                review_dict['ease'] = ease
                review_dict['satisfaction'] = satisfaction
                #review_dict['comment'] = comment

                # Add the review dictionary to the csv file
                writer.writerow(review_dict.values())
                print(review_dict)
            
            # Click 'Next' button
            try:
                time.sleep(2)
                next_button = driver.find_element_by_link_text('Next')
                next_button.click()
            except Exception as e:
                print(e)
                
    # Close File and Driver
    csv_file.close()
    driver.close()       
    
##WARNING THIS SELENIUM OF WEBMD TAKES VERY LONG. TOOK ME ABOUT 3 DAYS TO SCRAPE 93 DRUG REVIEWS!! DO AT YOUR OWN CAUTION!##
```
