```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
import csv
import pandas as pd 


options = webdriver.ChromeOptions()
options.add_experimental_option('w3c', False)

driver = webdriver.Chrome(executable_path=r'C:\Users\hank\Downloads\chromedriver_win32 (1)\chromedriver.exe',options=options)

review_dict = []    
for URL in URL_list:
    driver.get(URL)
    wait_review = WebDriverWait(driver, 10)

    while True:
        reviews = wait_review.until(EC.presence_of_all_elements_located((By.XPATH,'//div[@class="ddc-comment"]')))
        for review in reviews:

            drug_name = driver.find_element_by_xpath('//div[@class="contentBox"]/h1').text[17:]
            dates = review.find_element_by_xpath('.//span[@class="comment-date text-color-muted"]').text
            print(drug_name)
            print(dates)
            new = ((drug_name, dates))
            review_dict.append(new)    

            # At the end of the page go to next page
        try:
            next_button = wait_review.until(EC.element_to_be_clickable((By.LINK_TEXT, "Next")))
            next_button.click()
        except Exception as e:
                break
df = pd.DataFrame(review_dict, columns = ['drug_name', 'dates'])
driver.close()


            
```
