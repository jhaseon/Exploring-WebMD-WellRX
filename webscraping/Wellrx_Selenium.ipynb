```python
# Import necessary modules
from selenium import webdriver
import time
import re
import time 
import csv 

# Selenium
# Change your file path to where your chromedriver.exe is 
driver = webdriver.Chrome(r'..\chromedriver_win32\chromedriver.exe') #Filepath for webdriver

# Create an csv file with column names
with open("Wellrx_Selenium.csv" , 'w', encoding = 'utf-8', newline = '') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['drug name', 'price', 'other name', 'price1'])
    
    # Will Run Selenium for each URL page 
    for URL in URL_list:
        driver.get(URL)
        time.sleep(5)

        # Create xpath variable for individual drug page
        pages = driver.find_elements_by_xpath('//section[@class="main-section"]')
        for page in pages:
            
            # Each review will its own dictionary
            # That dictionary will be saved to csv file
            pages_dict = {}
    
            # Get drug name and price 
            drug_name = driver.find_element_by_xpath('//h4[@class="displayName skel skel-displayName"]').text 
            price = driver.find_element_by_xpath('//p[@class="right pr2"]').text
            pages_dict['drug name'] = drug_name
            pages_dict['price'] = price  
            
            # Click the alternative name 
            try:    
                name = driver.find_element_by_xpath('//h5[@id="OtherName"]').text
                name1 = re.findall(": (.*)", name)[0]
                driver.find_element_by_link_text(name1).click()
                time.sleep(5)
                
            # If there is no alternative name to click then add straight to csv
            except IndexError as error:
                writer.writerow(pages_dict.values())
            
            # If there is an alternative name, get other drug name and price
            else:
                other_name = driver.find_element_by_xpath('//h4[@class="displayName skel skel-displayName"]').text 
                price1 = driver.find_element_by_xpath('//p[@class="right pr2"]').text
                pages_dict['other name'] = other_name
                pages_dict['price1'] = price1
                
                # Add to csv
                writer.writerow(pages_dict.values())
                
# Close csv and driver
csv_file.close()
driver.close() 


```
