```python
import pandas as pd 
```

```python
# Import dataset 

df_WebMD = pd.read_csv('Webmd_and_Wellrx.csv')
df_drugs = pd.read_csv('drug_dates.csv')
```

```python
# Data Extraction -> From Reviewer columns -> ['age', 'Time', 'Reviewer_status', "Gender"]
df_WebMD["age"] = df_WebMD["Reviewer"].str.extract('(13-18|19-24|25-34|35-44|45-54|55-64|65-74|75 or over)')
df_WebMD["Time"] = df_WebMD["Reviewer"].str.extract('(less than 1 month|1 to 6 months|6 months to less than 1 year|1 to less than 2 years|2 to less than 5 years|5 to less than 10 years|10 years or more)')
df_WebMD["Reviewer_status"] = df_WebMD["Reviewer"].str.extract('(Patient|Caregiver)')
df_WebMD["Gender"] = df_WebMD["Reviewer"].str.extract('(Male|Female)')
```

```python
# Gender percentages and max 
df_WebMD_Gender = df_WebMD.groupby(['Gender']).agg({'Gender': 'count'}).rename(columns={'Gender': 'total'}).reset_index()
print(round((df_WebMD_Gender['total'].max()/df_WebMD_Gender['total'].sum()), 2))
df_WebMD_Gender
```

```python
# Age percentages and max 
df_WebMD_age = df_WebMD.groupby(['age']).agg({'age': 'count'}).rename(columns={'age': 'total'}).reset_index()
print(round((df_WebMD_age['total'].max()/df_WebMD_age['total'].sum()), 2))
df_WebMD_age
```

```python
#Length of Review percentages and max 
df_WebMD_Time = df_WebMD.groupby(['Time']).agg({'Time': 'count'}).rename(columns={'Time': 'total'}).reset_index()
print(round((df_WebMD_Time['total'].max()/df_WebMD_Time['total'].sum()), 2))
df_WebMD_Time
```

```python
# Reviewer status percentages and max 
df_WebMD_Reviewer_status = df_WebMD.groupby(['Reviewer_status']).agg({'Reviewer_status': 'count'}).rename(columns={'Reviewer_status': 'total'}).reset_index()
print(round((df_WebMD_Reviewer_status['total'].max()/df_WebMD_Reviewer_status['total'].sum()), 2))
df_WebMD_Reviewer_status
```
